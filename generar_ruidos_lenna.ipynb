```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.util import random_noise
```

```python
def plot_images(image, noisy_image, noise_type):
    """
    Crea un gr√°fico comparativo de la imagen ruidosa, la original y las versiones sin ruido corregidas tanto con la
    media como con la mediana.
    """

    for i, title, img in zip(range(1, 5),
                             ['Original', 'Ruido '+noise_type],
                             [image, noisy_image]):
        plt.subplot(1, 2, i)
        plt.imshow(img, cmap='gray')
        plt.title(f'{title}')
        plt.axis('off')
    plt.show()
    plt.close()
```

```python
image = imread("readmeImages/lenna.jpg")
```

```python
# Gaussiano
noisy_image = random_noise(image, 'gaussian', var=0.2 ** 2)
plot_images(image, noisy_image, 'Gaussiano')
```

```python
# Salt & pepper
noisy_image = random_noise(image, 's&p', amount=0.2)
plot_images(image, noisy_image, 'Salt & Pepper')
```

```python
# Uniforme
amplitud = 0.2
noise = np.random.uniform(low=0, high=amplitud, size=image.shape)
# Asegura que los valores est√©n en el rango correcto
image_flt = image.astype(float) / 255
noise = np.clip(noise, 0, 1)

# Agrega el ruido a la imagen
noisy_image = np.clip(image_flt + noise, 0, 1)
plot_images(image, noisy_image, 'Uniforme')
```
